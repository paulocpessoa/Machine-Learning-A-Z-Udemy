```python
# Apriori

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Data Preprocessing
dataset = pd.read_csv('C:\git\Machine-Learning-A-Z-Udemy\data_files\Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, len(dataset)):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
# Training Apriori on the dataset
from apyori import apriori
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

```python
# Visualising the results
results = list(rules)
```

```python
lst = []
for i in range(0, len(results)):
    if 'nan' not in list(results[i].ordered_statistics[0].items_base):
        lst.append({'Support': results[i].support,
                    'Base': results[i].ordered_statistics[0].items_base,
                    'items_add': results[i].ordered_statistics[0].items_add,
                    'confidence': results[i].ordered_statistics[0].confidence,
                    'lift': results[i].ordered_statistics[0].lift}
                  )
        
resultados_df = pd.DataFrame(lst)
resultados_df.head(10)
```

```python
resultados_df.sort_values(by='lift',ascending=False)
```

```python
resultados_df.sort_values(by='confidence',ascending=False)
```

```python
# Random Selection

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('C:\git\Machine-Learning-A-Z-Udemy\data_files\Ads_CTR_Optimisation.csv')

# Implementing Random Selection
import random
N = 10000
d = 10
ads_selected = []
total_reward = 0
```

```python
for n in range(0, N):
    random_ad = random.randrange(d)
    ads_selected.append(random_ad)
    reward = dataset.values[n, random_ad]
    total_reward = total_reward + reward
```

```python
for n in range(0, N):
    ad = random.randrange(d)
    ads_selected.append(ad)
    reward = dataset.values[n, ad]
    total_reward = total_reward + reward
```
